```python
import numpy as np
import matplotlib.pyplot as plt

# Function to minimize
def f(x):
    return x**2

# Gradient of the function
def grad_f(x):
    return 2 * x
```

```python
def gradient_descent_visualize(start_x=8, learning_rate=0.2, steps=12):
    x = start_x

    x_vals = np.linspace(-10, 10, 400)
    y_vals = f(x_vals)

    plt.ion()
    plt.figure(figsize=(8, 6))

    for i in range(steps):
        plt.clf()

        # Plot the function curve
        plt.plot(x_vals, y_vals, label="f(x) = x²")

        # Plot current point
        y = f(x)
        plt.scatter(x, y)

        # Add label to the point
        plt.annotate(
            f"Step {i+1}\nx={x:.4f}",
            (x, y),
            textcoords="offset points",
            xytext=(10,10),
            arrowprops=dict(arrowstyle="->")
        )

        plt.title("Gradient Descent Visualization")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.legend()
        plt.grid(True)

        # Gradient Descent update
        x = x - learning_rate * grad_f(x)

        plt.pause(0.6)

    plt.ioff()
    plt.show()

    # Final result
    print("Final Results:")
    print(f"Minimum x ≈ {x}")
    print(f"Minimum f(x) ≈ {f(x)}")
```

```python
# Run visualization
gradient_descent_visualize()
```
